```python
import folium as fm 
import pandas as pd
# library for creating map
```

```python
my_map=fm.Map(location=[13.546541,16.57474], zoom_start=2) # to pin point the location
my_map
```

```python
cities=pd.read_csv('C:/Users/chetn/Downloads/Cities.csv')
```

```python
cities
```

```python
cities.head(5)
```

```python
city=cities.loc[0]
fm.Marker(
    location=[city['latitude'],city['longitude']]).add_to(my_map)

my_map

for i, city in cities.iterrows(): # iterrow give the row info which is in city objeect
    fm.Marker(location=[city['latitude'],city['longitude']], popup=city['city'], tooltip=city['city']).add_to(my_map)

my_map

# popup - used to show name given to map on clicking
#tooltip - ussed to show name when we hover over it on map
```

```python
def select_marker_color(row):
    if row ['temperature']>= 12:
        return 'red'
    else:
        return 'blue'
    return 'green'
```

```python
cities['color']= cities.apply(select_marker_color, axis=1)
cities.head(10)
```

```python
my_map=fm.Map(location=[13.546541,16.57474], zoom_start=2) # to pin point the location
my_map

city=cities.loc[0]
fm.Marker(
    location=[city['latitude'],city['longitude']]).add_to(my_map)

my_map

for i, city in cities.iterrows(): # iterrow give the row info which is in city objeect
    fm.Marker(location=[city['latitude'],city['longitude']], popup=city['city'], 
              tooltip=city['city'], icon=fm.Icon(color=city['color'],prefix='fa',icon='circle')).add_to(my_map)

my_map
```

```python
my_map.save('my_awesm_map.html')
```
